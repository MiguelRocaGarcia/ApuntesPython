## Group by

```python
#Hace agrupaciones de una lista según lo que devuelva una función
from itertools import groupby

def mayorQue3(x):
  return x > 3

a = [1,2,3,4,5]
group_obj = groupby(a, key=mayorQue3)

for key, value in group_obj:
  print(key, list(value))
```

#### Herencia

```python
class ClasePadre:  
  def __init__(self, atributoPadre1, atributoPadre2):
    self.atributoPadre1 = atributoPadre1
    self.atributoPadre2 = atributoPadre2
  def metodoPapa(self):
    print("Metodo del papa")
```

```python
class ClaseHija(ClasePadre):
  def __init__(self, atributoPadre1, atributoPadre2, atributoHija1, atributoHija2):
    ClasePadre.__init__(self, atributoPadre1,atributoPadre2)
    self.atributoHija1 = atributoHija1
    self.atributoHija2 = atributoHija2
```

```python
clasePapa = ClasePadre("AtributoPapa1", "AtributoPapa2")
claseHija = ClaseHija(atributoHija1=1,atributoHija2=2,atributoPadre1=3,atributoPadre2=4)
claseHija.metodoPapa()
```

```python
#Expresiones Regulares
# .     Un caracter cualquiera
# a*    Caracter repetido de 0 a infinito
# a+    Caracter repetido de 1 a infinito veces
# a?    Una o ninguna repetición
# Para referirnos a un caracter especial hay que escaparlo con barra invertida \
# \d    Cualquier dígito decimal
# \w    Cualquier caracter alfanumérico
# \n    Salto de línea
# \t    Tabulador
# \s    Cualquier caracter de espaciado: Espacio, Tabulador, Salto de línea, etc.
# [^a] Cualquier caracter excepto a
# ^ab   Empieza por a
# ab$   Termina por b


# '.*c' aplicada a 'ababcababcab' devolverá 'ababcababc' en lugar de 'ababc'
# Para hacer que el emparejamiento no sea voraz, es decir, que busque todos los emparejamientos, se usan *?, +? y ??. Sino, devuelve hasta la aparicion más lejana
# '.*?c' aplicada a 'ababcababcab' devolverá 'ababc'
#Habrá que poner x*?, x+?, x??

# a|b   Funciona como un OR
# a{n}  n repeticiones exactas
# a{n,m} Entre n y m repeticiones
# (ab)* Se puede agrupar con paréntesis. En los métodos de re devolverá el contenido de los paréntesis
# [aeiou] Conjunto de caracteres. Que contenga cualquiera de los caracteres entre los corchetes una vez

# [0-9] Cualquier número del rango del 0 al 9
# Otros rangos: [a-z], [A-Z], [A-z], [a-zA-Z0-9]

# Página para probar expresiones regulares: https://regex101.com/
```

### SQL

Sólo por curiosidad, el módulo pandas.io.sql proporciona una colección de wrappers para facilitar la recuperación de datos y reducir la dependencia de la API específica de la base de datos. La abstracción de la base de datos es proporcionada por SQLAlchemy, así que tiene que estar instalado junto con las librerías de los motores concretos de las bases de datos, por ejemplo psycopg2 para PostgreSQL o pymysql para MySQL. En el caso de SQLite, está incluido en la biblioteca estándar de Python de forma predeterminada.

Las funciones clave son:

* `read_sql_table(nombre_tabla, con[, schema, ...])`	
    * Cargar la tabla en un DataFrame.
* `read_sql_query(sql, con[, index_col, ...])`
    * Cargar una consulta concreta en un DataFrame.
* `read_sql(sql, con[, index_col, ...])`
    * Un wrapper de ambos (consulta o tabla).
* `DataFrame.to_sql(name, con[, flavor, ...])`
    * Escribe los registros del DataFrame a una tabla de la base de datos.

```python
import logging
import threading
import time

def thread_function(name):
    logging.info("Thread %s: starting", name)
    time.sleep(2)
    logging.info("Thread %s: finishing", name)

if __name__ == "__main__":
    format = "%(asctime)s: %(message)s"
    logging.basicConfig(format=format, level=logging.INFO,
                        datefmt="%H:%M:%S")

    threads = list()
    for index in range(3):
        logging.info("Main    : create and start thread %d.", index)
        x = threading.Thread(target=thread_function, args=(index,))
        threads.append(x)
        x.start()

    for index, thread in enumerate(threads):
        logging.info("Main    : before joining thread %d.", index)
        thread.join()
        logging.info("Main    : thread %d done", index)
```

```python
import logging
import threading
import time
import random

# The 'results' list will hold our retrieved data
# The 'urls' list contains all of the urls that are to be checked for data


def algo(parametro1, parametro2):
    time.sleep(random.uniform(1, 3))
    print( str(parametro1) + " " + str(parametro2))


#create a list of threads
threads = []

for i in range(10):
    process = threading.Thread(target=algo, args=[i,7])
    process.start()
    threads.append(process)
# We now pause execution on the main thread by 'joining' all of our started threads.
# This ensures that each has finished processing the urls.
for process in threads:
    process.join()
# At this point, results for each URL are now neatly stored in order in 'results'
print("Se acabó")
```

```python
import concurrent.futures
import urllib.request

URLS = ['http://www.foxnews.com/',
        'http://www.cnn.com/',
        'http://europe.wsj.com/',
        'http://www.bbc.co.uk/',
        'http://some-made-up-domain.com/']

# Retrieve a single page and report the URL and contents
def load_url(url, timeout):
    with urllib.request.urlopen(url, timeout=timeout) as conn:
        return conn.read()

# We can use a with statement to ensure threads are cleaned up promptly
with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor:
    # Start the load operations and mark each future with its URL
    future_to_url = {executor.submit(load_url, url, 60): url for url in URLS}
    for future in concurrent.futures.as_completed(future_to_url):
        url = future_to_url[future]
        try:
            data = future.result()
        except Exception as exc:
            print('%r generated an exception: %s' % (url, exc))
        else:
            print('%r page is %d bytes' % (url, len(data)))
```

```python
#Scraper de LinkedIn 
"""
Mediante la terminal de comandos, he instalado los sigueintes paquetes:
pip3 install selenium
pip3 install bs4
"""

from selenium import webdriver
from bs4 import BeautifulSoup as bs
import re, random, time, json
import credentials


#Abrimos LinkedIn en google chrome mediante su controlador
def startChrome():
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument("--incognito") #Modo incognito
    driver = webdriver.Chrome(credentials.CHROMEDRIVER, options=chrome_options)
    driver.get('https://www.linkedin.com/login/es?fromSignIn=true&trk=guest_homepage-basic_nav-header-signin')
    return driver

#Iniciar sesión
def log_in(driver, log_username, log_password):
    username = driver.find_element_by_id('username')
    username.send_keys(log_username)
    password = driver.find_element_by_id('password')
    password.send_keys(log_password)
    log_in_button = driver.find_element_by_xpath('//*[@type="submit"]')
    log_in_button.click()

#Bajar al fondo de la página
def scroll_down(driver):
    new_height = 0
    old_height = 0
    cont=0
    total_height = driver.execute_script("return document.body.scrollHeight")
    while new_height < total_height:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  # Scroll down to bottom
        time.sleep(random.uniform(2.9, 4.9))  # Wait to load page
        new_height = driver.execute_script("return document.body.scrollHeight")  # Calculate new scroll height and compare with total scroll height
        if(old_height == new_height and cont>=3):
            break
        old_height = new_height
        cont += 1


#Obtener URLs de los contactos
def getContactsURLs(driver):
    driver.get('https://www.linkedin.com/mynetwork/invite-connect/connections/')
    time.sleep(1)
    scroll_down(driver)
    contacts_find = bs(driver.page_source, features="html.parser")
    contacts_urls = contacts_find.find_all('a', class_="mn-connection-card__link ember-view")
    mynetwork = []
    for url_contact in contacts_urls:
        mynetwork.append(url_contact.get('href'))
    return  mynetwork

#Abrir los ver más...
def extend_see_more(driver):  #Extendemos los ver más
    extend_buttons = driver.find_elements_by_xpath('// *[ @ id = "line-clamp-show-more-button"]')   #en el acerca de...
    extend_buttons.extend(driver.find_elements_by_xpath('/html/body/div[8]/div[3]/div/div/div/div/div[2]/main/div[2]/div[5]/span/div/section/div[1]/section/div/button')) #El ver más de experiencia
    extend_buttons.extend(driver.find_elements_by_xpath('/html/body/div[8]/div[3]/div/div/div/div/div[2]/main/div[2]/div[5]/span/div/section/div[1]/section/ul/li[2]/section/div[2]/div[2]/button')) #El ver más de experiencia
    for button in extend_buttons:
        button.send_keys("\n")

#Experiencia
def get_experience(datos, contact_page):    # Experiencia
    try:
        exp_section = contact_page.find('section', {'id': 'experience-section'})
        exp_section = exp_section.find('ul')
        exp_list = exp_section.find_all('li')
        experience = {}
        datos['Experience'] = []
        for li_exp in exp_list:
            try:
                a_tag = li_exp.find('div').find('a')
                experience['Position'] = a_tag.find('h3').get_text().strip()
                experience['Company'] = a_tag.find_all('p')[1].get_text().strip()
                experience['Period'] = a_tag.find_all('h4')[0].find_all('span')[1].get_text().strip()
                experience['Time'] = a_tag.find_all('h4')[1].find_all('span')[1].get_text().strip()
                #employment_type
                #location
                #description
                #Si pone localizacion, no va

            except:
                a_tag = li_exp.find('div').find('a')
                if (a_tag):
                    experience = {}
                    experience['Company'] = a_tag.find_all('h3')[0].find_all('span')[1].get_text().strip()
                    experience['Time'] = a_tag.find_all('h4')[0].find_all('span')[1].get_text().strip()

                    li_position = li_exp.find('ul').find_all('li')
                    experience['Positions'] = []
                    pos = {}
                    for position in li_position:
                        pos['Position'] = position.find_all('h3')[0].find_all('span')[1].get_text().strip()
                        pos['Period'] = position.find_all('h4')[0].find_all('span')[1].get_text().strip()
                        pos['Time'] = position.find_all('h4')[1].find_all('span')[1].get_text().strip()
                        experience['Positions'].append(pos.copy())

            datos['Experience'].append(experience.copy())
    except:
        pass

#Información básica: Nombre, Trabajo, Localización, etc.
def get_basic_info(datos, contact_page):
    # Nombre
    name = contact_page.find('li', class_="inline t-24 t-black t-normal break-words")
    datos['Name'] = name.contents[0].strip()
    # Trabajo
    job = contact_page.find('h2', class_='mt1 t-18 t-black t-normal break-words')
    datos['Job'] = job.contents[0].strip()
    # Localización
    location = contact_page.find('li', class_='t-16 t-black t-normal inline-block')
    datos['Location'] = location.contents[0].strip()
    # Número de contactos
    num_contacts = contact_page.find('a', {'data-control-name': 'topcard_view_all_connections'})
    if (num_contacts):
        num_contacts = num_contacts.find('span')
        datos['Number_of_Contacts'] = num_contacts.contents[0].strip().split()[0]
    else:
        num_contacts = contact_page.find('span', class_='t-16 t-black t-normal')
        if (num_contacts):
            datos['Number_of_Contacts'] = num_contacts.contents[0].strip()

#Información de contacto
def contact_info(url, datos):
    driver.get(url + 'detail/contact-info/')    # Accedemos a la página de información de contacto
    contact_page_info = bs(driver.page_source, features="html.parser")
    # Email
    email = contact_page_info.find('a', href=re.compile("mailto"))
    if (email):
        datos['email'] = email.get('href')[7:]
    # Fecha de conexión
    connection_date = contact_page_info.find('section', class_='pv-contact-info__contact-type ci-connected')
    if (connection_date):
        connection_date = contact_page_info.find('span', class_='pv-contact-info__contact-item')
        datos['Connection_date'] = connection_date.contents[0].strip()
    # Teléfonos
    phone_number = contact_page_info.find('section', class_='pv-contact-info__contact-type ci-phone')
    if (phone_number):
        phone_number = phone_number.find(class_='pv-contact-info__ci-container').find('span')
        datos['Phone_numbers'] = phone_number.contents
    # Enlace LinkedIn
    datos['LinkedIn'] = url


#OBTENCIÓN DE INFORMACIÓN DE UN USUARIO
def userInfo(driver, url):
    datos = {}
    driver.get(url)
    scroll_down(driver)
    extend_see_more(driver)
    contact_page = bs(driver.page_source, features="html.parser")
    time.sleep(1)
    get_basic_info(datos, contact_page) #Información básica: Nombre, Trabajo, Localización, etc.
    get_experience(datos, contact_page) #Experiencia
    contact_info(url, datos)    #Información de contacto
    return datos



#Main
driver = startChrome()  #Abrir el browser y linkedin.com
log_in(driver, credentials.LOGIN_USERNAME, credentials.LOGIN_PASSWORD)    #Inicio de sesión
if(driver.current_url != 'https://www.linkedin.com/feed/?trk=guest_homepage-basic_nav-header-signin'):  #Por si sale el control de bots
    print("Rellena los campos, pasa el control e introduce un caracter: ")
    _ = input()
my_network = getContactsURLs(driver)    #Obtención de la url de los contactos
connectionsJSON = open('LinkedIn_Connections.json', 'w')
infoConnections = {'Contact': []}

#my_network[0] = '/in/miguel-roca-garcía-9908a6200/'
i=0
for contact_url in my_network:
    infoConnections['Contact'].append(userInfo(driver, 'https://www.linkedin.com' + contact_url))   #Añadimos al diccionario la info de cada usuario por cada url
    print(infoConnections['Contact'][-1])
    i+=1
    print(i, '/', len(my_network))

json.dump(infoConnections, connectionsJSON, indent=4)
connectionsJSON.close()
driver.quit()

#Quitar limite de 3 usuarios. Quitar mi usuario


"""

Experiencia: pv-entity__position-group-pager pv-profile-section__list-item ember-view
pv-profile-section__card-item-v2 pv-profile-section pv-position-entity ember-view

Educación: pv-profile-section__list-item pv-education-entity pv-profile-section__card-item ember-view
"""

"""
#pip3 install parsel
from parsel import Selector
sel = Selector(text=driver.page_source)
name = sel.xpath('//*[starts-with(@class, "inline t-24 t-black t-normal break-words")]/text()').extract_first()
name = name.strip()
job_title = sel.xpath('//*[starts-with(@class, "mt1 t-18 t-black t-normal break-words")]/text()').extract_first()
job_title = job_title.strip()
location = sel.xpath('//*[starts-with(@class, "t-16 t-black t-normal inline-block")]/text()').extract_first()
location = location.strip()"""
```
